```python
import pandas as pd
import os
import glob

# Define the subfolder path
subfolder_path = './data/segSummaries/'

# Use glob to get all the files ending with "_Summary_SEGID.csv"
file_paths = glob.glob(os.path.join(subfolder_path, "*.csv"))

# Create an empty list to store dataframes
dfs = []

# Loop through the files, parse filename, and read them into dataframes
for file_path in file_paths:
    # Get filename without extension
    filename = os.path.splitext(os.path.basename(file_path))[0]
    
    # Split filename based on double underscore
    mdlVersion, scnGroup, scnYear = filename.split('__')
    
    # Read the CSV into a dataframe
    df = pd.read_csv(file_path)
    
    # Add new columns with values from filename
    df['mdlVersion'] = mdlVersion
    df['scnGroup'] = scnGroup
    df['scnYear'] = scnYear
    
    dfs.append(df)

# If you want to concatenate all the dataframes into a single dataframe
dfCombined = pd.concat(dfs, ignore_index=True)
dfCombined['scnYear'] = dfCombined['scnYear'].astype('int64')
dfCombined
```

```python
dfCombined_melt = dfCombined.melt(id_vars=['SEGID','mdlVersion','scnGroup','scnYear'], value_vars=dfColKeys['col'].tolist(), var_name='col', value_name='val')

dfCombined_melt
```

```python
dfCombined_melt_withkeys = pd.DataFrame.merge(dfCombined_melt, dfColKeys, on='col')
dfCombined_melt_withkeys
```

```python
[col for col in dfCombined_melt_withkeys.columns if col != 'att']
```

```python
dfCombined_melt_withkeys_pivot_atts = dfCombined_melt_withkeys.pivot(index=[col for col in dfCombined_melt_withkeys.columns if col not in ['att','val','col']], columns="att", values="val")
dfCombined_melt_withkeys_pivot_atts = dfCombined_melt_withkeys_pivot_atts.reset_index()
#dfCombined_melt_withkeys_pivot_atts = dfCombined_melt_withkeys_pivot_atts.fillna('blank')
dfCombined_melt_withkeys_pivot_atts
```

```python
# Base directory where folders are created
base_dir = './_site/data/scnData/'

_df = dfCombined_melt_withkeys_pivot_atts

# Loop through scenarios in dfScn
for _, row in dfScn.iterrows():
    print(row['mdlVersion'] + row['scnGroup'] + str(row['scnYear']))
    # Filter the main dataframe based on the scenario
    filtered_df = _df[
        (_df['mdlVersion'] == row['mdlVersion']) & 
        (_df['scnGroup'] == row['scnGroup']) & 
        (_df['scnYear'] == row['scnYear'])
    ].copy()
    
    # Create the filename based on the scenario and directory structure
    folder_name = f"{row['mdlVersion']}__{row['scnGroup']}__{row['scnYear']}"
    full_path = os.path.join(base_dir, folder_name, 'roadsegsummary.json')

    filtered_df.drop(columns=['mdlVersion','scnGroup','scnYear'], inplace=True)

    # Export the filtered dataframe to a json file
    filtered_df = filtered_df.rename(columns={'SEGID':'id'})
    filtered_df.to_json(full_path, orient='records', lines=True)
```
